```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
listaBeacon = np.loadtxt("BeaconName.txt", skiprows = 2)
```

```python
print(listaBeacon)
```

```python
plt.figure(figsize = (10,8))
plt.title("Medicion Beacon con Interferencias")
plt.xlabel("Mediciones")
plt.ylabel("Intensidad (dBm)")
plt.grid()
plt.plot(listaBeacon, '-', color="r",
             label="Mediciones Beacon")
plt.legend(loc="best")
plt.show()
```
