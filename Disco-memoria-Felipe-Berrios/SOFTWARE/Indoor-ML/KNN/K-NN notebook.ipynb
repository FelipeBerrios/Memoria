# KNN

## Dataset Completo

```python
# import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn_porter import Porter
```

```python
Xdf_raw = pd.read_csv("../IndoorFingerprint.csv").drop("Y", axis=1)
Ydf_raw = pd.read_csv("../IndoorFingerprint.csv").drop("X", axis=1)
```

## Modelo X

```python
# Se normalizan los datos para ambos
atribute_numbers = Xdf_raw.shape[1]-1
scaler = preprocessing.StandardScaler()
Xdf_raw.iloc[:, 1:atribute_numbers+1] = scaler.fit_transform(Xdf_raw.iloc[:, 1:atribute_numbers+1])
X_std = Xdf_raw.iloc[:, 1:atribute_numbers+1]
y = Xdf_raw.iloc[:,0]
```

```python
# Utilizado para guardar datos
print(scaler.scale_)
print(scaler.mean_)
scale = np.array(scaler.scale_)
mean = np.array(scaler.mean_)

np.savetxt('scale.txt',[scale] , delimiter=' ', fmt='%.8f')
np.savetxt('mean.txt',[mean] , delimiter=' ', fmt='%.8f')
np.savetxt('original_scaled.txt', X_std, delimiter=' ', fmt='%.8f' )
```

```python
# KNN
k = 2
clf = KNeighborsClassifier(n_neighbors=2)
clf.fit(X_std, y)
```

```python
porterKNN = Porter(clf, language='java')
output = porterKNN.export()
print(output)
```

## Modelo Y

```python
# Se normalizan los datos para ambos
atribute_numbers = Xdf_raw.shape[1]-1
scaler = preprocessing.StandardScaler()
Ydf_raw.iloc[:, 1:atribute_numbers+1] = scaler.fit_transform(Ydf_raw.iloc[:, 1:atribute_numbers+1])
X_std_y = Ydf_raw.iloc[:, 1:atribute_numbers+1]
y_y = Ydf_raw.iloc[:,0]
```
